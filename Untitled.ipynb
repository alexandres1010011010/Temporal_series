## Produto mais popular de uma determinada loja

```python
df1.info()
```

```python
df1["StoreID"].unique()
```

```python
id_loja = df1["StoreID"].unique()[0]
dataframe = df1
filtrando = dataframe[dataframe["StoreID"] == id_loja]
filtrando
```

```python
filtrando = filtrando[["ProductID", "Discount", "UnitPrice", "Quantity"]]
filtrando
```

```python
agrupando = filtrando.groupby("ProductID").sum()
agrupando
```

```python
sorteando_quantidade = agrupando.sort_values(by="Quantity", ascending=False)
```

```python
sorteando_quantidade
```

```python
produto_mais_popular_por_quantidade = sorteando_quantidade.head(1)
produto_mais_popular_por_quantidade
```

```python
def produto_popular(id_loja, dataframe):
    filtrando = dataframe[dataframe["StoreID"] == id_loja]
    filtrando = filtrando[["ProductID", "Discount", "UnitPrice", "Quantity"]]
    agrupando = filtrando.groupby("ProductID").sum()
    sorteando_quantidade = agrupando.sort_values(by="Quantity", ascending=False)
    sorteando_preco = agrupando.sort_values(by="UnitPrice", ascending=False)
    produto_mais_popular_por_quantidade = sorteando_quantidade.head(1)
    produto_mais_popular_por_preco = sorteando_preco.head(1)
    print(f"O produto mais popular por quatidade é {produto_mais_popular_por_quantidade} da loja {id_loja}!")
    print(f"O produto mais popular por preço é {produto_mais_popular_por_preco} da loja {id_loja}!")    
for i in df1["StoreID"].unique():
    produto_popular(i, df1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_mensalmente[['Date', 'Discount', 'UnitPrice']], df_mensalmente['Quantity'], test_size=0.2, random_state=42)
```

```python
df_mensalmente = df_mensalmente[["Date", "Discount", "UnitPrice", "Quantity","Month"]]
df_mensalmente
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

#df_mensalmente['Year'] = df_mensalmente['Date'].dt.year
df_mensalmente['Month'] = df_mensalmente['Date'].dt.month

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_mensalmente[['Month', 'Discount', 'UnitPrice']], df_mensalmente['Quantity'], test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions)

# Print evaluation metrics
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
```
