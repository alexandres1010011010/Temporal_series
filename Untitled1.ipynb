```python
df["Date"] = pd.to_datetime(df["Date"])
df["Discount"] = df["Discount"].replace(',', '.', regex=True)
df["Discount"] = pd.to_numeric(df["Discount"])
df["UnitPrice"] = df["UnitPrice"].replace(',', '.', regex=True)
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"])
df["Quantity"] = pd.to_numeric(df["Quantity"])
df["ClientID"] = pd.to_numeric(df["ClientID"])
df["StoreID"] = pd.to_numeric(df["StoreID"])
#df["ID"] = pd.to_numeric(df["ID"])
#df["ProductID"] = pd.to_numeric(df["ProductID"])
```

```python
# Assuming your sales table is named 'sales_df'

# Group by the product ID and calculate aggregate statistics
grouped_df = df.groupby('ProductID').agg({
    'Quantity': 'sum',
    'Discount': 'mean',
    'UnitPrice': 'mean',
    'ID': 'count'
})

# Rename the columns for clarity
grouped_df = grouped_df.rename(columns={
    'Quantity': 'TotalQuantity',
    'Discount': 'MeanDiscount',
    'UnitPrice': 'MeanUnitPrice',
    'ID': 'TotalSales'
})

```

```python
import pandas as pd

# Assuming your sales data is in a DataFrame named 'sales_data' with columns: 'StoreID', 'Date', and 'SalesRevenue'

# Group the sales data by store and date, and calculate monthly sales
monthly_sales = df_novo.groupby(['StoreID', pd.Grouper(key='Date', freq='M')])['receita liquida'].sum().reset_index()

# Calculate outliers for each store's monthly sales
outliers = pd.DataFrame()
for store_id, store_group in monthly_sales.groupby('StoreID'):
    q1 = store_group['receita liquida'].quantile(0.25)
    q3 = store_group['receita liquida'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    store_outliers = store_group[(store_group['receita liquida'] < lower_bound) | (store_group['receita liquida'] > upper_bound)]
    outliers = pd.concat([outliers, store_outliers])

# Print the outliers for each store's monthly sales
print(outliers)
```
